```python
means = [ghz26_1shard40[ghz26_1shard40['keylength'] > 10]['dif'].mean(), ghz26_shard40[ghz26_shard40['keylength'] > 10]['dif'].mean(), ghz26_shard80[ghz26_shard80['keylength'] > 10]['dif'].mean()]
ninefive = [np.percentile(ghz26_1shard40[ghz26_1shard40['keylength'] > 10]['dif'], 95),  np.percentile(ghz26_shard40[ghz26_shard40['keylength'] > 10]['dif'], 95),  np.percentile(ghz26_shard80[ghz26_shard80['keylength'] > 10]['dif'], 95)]
ninenine = [np.percentile(ghz26_1shard40[ghz26_1shard40['keylength'] > 10]['dif'], 99), np.percentile(ghz26_shard40[ghz26_shard40['keylength'] > 10]['dif'], 99),  np.percentile(ghz26_shard80[ghz26_shard80['keylength'] > 10]['dif'], 99)]
means_std = [np.std(ghz26_1shard40[ghz26_1shard40['keylength'] > 10]['dif']), np.std(ghz26_shard40[ghz26_shard40['keylength'] > 10]['dif']), np.std(ghz26_shard80[ghz26_shard80['keylength'] > 10]['dif'])]

labels = ['1x 40GB', '5x 8 GB', '5x 15 GB']


x = np.arange(len(labels))  # the label locations
width = 0.2  # the width of the bars

fig, ax = plt.subplots(figsize=(12,8), dpi=300)
rects1 = ax.bar(x - width, np.round(means, 0), width, label='Mean', yerr=means_std)
rects2 = ax.bar(x, np.round(ninefive, 0), width, label='95-percentile')
rects3 = ax.bar(x + width, np.round(ninenine, 0), width, label='99-percentile')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Service Time (ms)')
ax.set_title('Service Time by Number of Shards and its Physical Size')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)
autolabel(rects3)


#fig.tight_layout()

plt.show()
fig.savefig('D:\\elastic-hurryup\\graphs\\shards-size-comparison.png')
```
